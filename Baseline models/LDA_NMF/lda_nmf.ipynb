```python
aspects={}
for i,topic in enumerate(nmf.components_):
    print(f'Top 100 words for topic #{i}:')
    print([tfidf_vectorizer.get_feature_names()[i] for i in topic.argsort()[-100:]])
    print('\n')
    aspects[i]=[tfidf_vectorizer.get_feature_names()[i] for i in topic.argsort()[-100:]]
aspect_file = open('output_dir/aspect.log{}'.format(nmf.components_.shape[0]), 'wt', encoding='utf-8')
aspect_file.write(str(aspects))
aspect_file.close()
import pickle
with open('output_dir/aspect.log{}.pkl'.format(nmf.components_.shape[0]),'wb') as f:
          pickle.dump(aspects,f)
```

```python
# Initialize NMF
nmf = NMF(init='random', n_components=20, random_state=0)
W = nmf.fit_transform(tfidf)
H = nmf.components_
```

```python
W.shape, H.shape
```

```python
tfidf_vectorizer.vocabulary_.keys()
```

```python
terms = [""] * len(tfidf_vectorizer.vocabulary_)
terms
```

```python
for term in tfidf_vectorizer.vocabulary_.keys():
    terms[tfidf_vectorizer.vocabulary_[term]] = term
for topic_index in range(H.shape[0]):  # H.shape[0] is k
    top_indices = np.argsort(H[topic_index,:])[::-1][0:10]
    term_ranking = [terms[i] for i in top_indices]
    print("Topic {}: {}".format(topic_index, ", ".join(term_ranking)))
```
